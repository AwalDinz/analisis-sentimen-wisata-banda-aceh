Cleaning Text

```python
import re
```

```python
def clean_text(text):
    text = str(text)

    # hapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # hapus mention dan hashtag
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)

    # hapus angka
    text = re.sub(r'\d+', '', text)

    # hapus tanda baca dan karakter khusus
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df_clean = df_casefold.copy()

df_clean["text"] = df_clean["text"].apply(clean_text)

df_clean[["text"]].head()
```

```python
print("SEBELUM CLEANING")
print(df_casefold["text"].iloc[0])

print("\n" + "="*60 + "\n")

print("SESUDAH CLEANING")
print(df_clean["text"].iloc[0])
```

```python
print("Sebelum:", len(df_clean))

df_clean = df_clean[df_clean["text"].str.strip() != ""]

print("Sesudah:", len(df_clean))
```

```python
df_clean.to_csv(
    "dataset_clean.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Cleaning selesai")
```

Tokenizing

```python
df_token = df_clean.copy()

df_token["tokens"] = df_token["text"].apply(lambda x: x.split())

df_token[["text", "tokens"]].head()
```

```python
print(df_token["text"].iloc[0])

print("\n")

print(df_token["tokens"].iloc[0])
```

```python
df_token.to_csv(
    "dataset_tokenizing.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Tokenizing selesai")
```

Stopword Removal

```python
!pip install Sastrawi
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
factory = StopWordRemoverFactory()

stopwords = set(factory.get_stop_words())

print("Jumlah stopword :", len(stopwords))
```

```python
df_stopword = df_token.copy()

df_stopword["tokens"] = df_stopword["tokens"].apply(
    lambda words: [word for word in words if word not in stopwords]
)

df_stopword[["tokens"]].head()
```

```python
print("SEBELUM")
print(df_token["tokens"].iloc[0])

print("\n" + "="*60 + "\n")

print("SESUDAH")
print(df_stopword["tokens"].iloc[0])
```

```python
df_stopword.to_csv(
    "dataset_stopword.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Stopword removal selesai")
```

Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
df_stemming = df_stopword.copy()

df_stemming["tokens"] = df_stemming["tokens"].apply(
    lambda words: [stemmer.stem(word) for word in words]
)

df_stemming[["tokens"]].head()
```

```python
print("SEBELUM")
print(df_stopword["tokens"].iloc[0])

print("\n" + "="*60 + "\n")

print("SESUDAH")
print(df_stemming["tokens"].iloc[0])
```

```python
df_stemming["text_preprocessed"] = df_stemming["tokens"].apply(
    lambda x: " ".join(x)
)

df_stemming[
    ["text","text_preprocessed"]
].head()
```
