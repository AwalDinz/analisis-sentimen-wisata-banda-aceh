```python
import pandas as pd
import os

gmaps_path = "/content/drive/MyDrive/SEMESTER 6/DATASET/GOOGLE_MAPS"

all_data = []

for file in os.listdir(gmaps_path):
    if file.endswith(".csv"):
        try:
            df = pd.read_csv(
                os.path.join(gmaps_path, file),
                sep=';',
                encoding='latin1'
            )

            all_data.append(df)

        except Exception as e:
            print("ERROR :", file)
            print(e)

gmaps_df = pd.concat(all_data, ignore_index=True)

print("Jumlah baris :", gmaps_df.shape[0])
print("Jumlah kolom :", gmaps_df.shape[1])

gmaps_df.head()
```

```python
print(gmaps_df.columns.tolist())
```

```python
gmaps_clean = gmaps_df[
    ['title', 'text', 'stars', 'publishedAtDate', 'likesCount']
].copy()

gmaps_clean['platform'] = 'Google Maps'

print(gmaps_clean.shape)

gmaps_clean.head()
```

```python
print("Sebelum hapus:", len(gmaps_clean))

gmaps_clean = gmaps_clean.dropna(subset=['text'])

print("Sesudah hapus:", len(gmaps_clean))
```

```python
output_path = "/content/drive/MyDrive/SEMESTER 6/DATASET/google_maps_clean.csv"

gmaps_clean.to_csv(
    output_path,
    index=False,
    encoding='utf-8-sig'
)

print("Berhasil disimpan:", output_path)
```

```python
import os

tiktok_path = "/content/drive/MyDrive/SEMESTER 6/DATASET/TIKTOK"

files = os.listdir(tiktok_path)

print("Jumlah file:", len(files))
print(files)
```

```python
import pandas as pd
import os

tiktok_path = "/content/drive/MyDrive/SEMESTER 6/DATASET/TIKTOK"

all_tiktok = []

for file in os.listdir(tiktok_path):
    if file.endswith(".csv"):

        try:
            df = pd.read_csv(
                os.path.join(tiktok_path, file),
                sep=';',
                encoding='latin1'
            )
        except:
            df = pd.read_csv(
                os.path.join(tiktok_path, file),
                sep=',',
                encoding='latin1'
            )

        destinasi = file.replace("TikTok_", "").replace(".csv", "")

        df["title"] = destinasi

        all_tiktok.append(df)

tiktok_df = pd.concat(all_tiktok, ignore_index=True)

print("Jumlah baris:", tiktok_df.shape[0])
print("Jumlah kolom:", tiktok_df.shape[1])

tiktok_df.head()
```

```python
tiktok_clean = pd.DataFrame()

tiktok_clean["title"] = tiktok_df["title"]
tiktok_clean["text"] = tiktok_df["text"]

# TikTok tidak punya rating bintang
tiktok_clean["stars"] = None

# tanggal komentar
tiktok_clean["publishedAtDate"] = tiktok_df["createTimeISO"]

# jumlah like komentar
tiktok_clean["likesCount"] = tiktok_df["diggCount"]

# platform
tiktok_clean["platform"] = "TikTok"

print(tiktok_clean.shape)

tiktok_clean.head()
```

```python
print("Sebelum hapus:", len(tiktok_clean))

tiktok_clean = tiktok_clean.dropna(subset=["text"])

print("Sesudah hapus:", len(tiktok_clean))
```

```python
tiktok_clean.to_csv(
    "/content/drive/MyDrive/SEMESTER 6/DATASET/tiktok_clean.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Berhasil disimpan")
```

```python
final_df = pd.concat(
    [gmaps_clean, tiktok_clean],
    ignore_index=True
)

print("Total data:", final_df.shape)
```

```python
final_df.to_csv(
    "/content/drive/MyDrive/SEMESTER 6/DATASET/final_dataset_wisata_banda_aceh.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Dataset final berhasil disimpan")
```
